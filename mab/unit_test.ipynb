```python
import numpy as np
from mab.nonstationary_linear_bandits import SlidingWindowLinearTS

# Создаём два bandit с одинаковым seed
rng = np.random.default_rng(42)

# 1) Bandit с incremental update (your fixed version)
bandit_incr = SlidingWindowLinearTS(
    n_features=10,
    actions=[0, 1],
    window_size=50,
    lambda_prior=1.0,
    noise_std=0.01,
    seed=123,
    rebuild_interval=None,  # default = window_size
)

# 2) Bandit с forced full rebuild каждый update (старая логика)
bandit_full = SlidingWindowLinearTS(
    n_features=10,
    actions=[0, 1],
    window_size=50,
    lambda_prior=1.0,
    noise_std=0.01,
    seed=123,
    rebuild_interval=1,  # rebuild каждый update
)

# Запускаем 200 updates с одинаковыми данными
for t in range(200):
    x = rng.standard_normal(10)
    a = int(rng.random() < 0.5)
    r = rng.standard_normal() * 0.05

    bandit_incr.update(a, x, r)
    bandit_full.update(a, x, r)

# Сравниваем матрицы
for action in [0, 1]:
    A_diff = np.max(np.abs(bandit_incr.A[action] - bandit_full.A[action]))
    b_diff = np.max(np.abs(bandit_incr.b[action] - bandit_full.b[action]))
    print(f"Action {action}: max A diff = {A_diff:.2e}, max b diff = {b_diff:.2e}")

# Ожидаем разницу ~1e-12 или меньше (floating-point noise)
# Если разница > 1e-8 — есть bug
```
